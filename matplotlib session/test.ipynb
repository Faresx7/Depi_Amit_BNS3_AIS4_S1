```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/COMPUMARTS/Downloads/DEPI/Depi_Amit_BNS3_AIS4_S1/matplotlib session/train 1.csv')
df.tail(2)
```

```python
cols=['Age','Fare']
```

```python
plt.figure(figsize=(8,2))
for i , col in enumerate(cols):
    plt.subplot(1,2,i+1)
    plt.hist(df[col],edgecolor='white')
    plt.title(f'{col}')
```

```python
plt.figure(figsize=(8,2))
for i , col in enumerate(cols):
    plt.subplot(1,2,i+1)
    sns.kdeplot(df[col])
    plt.title(f'{col}')
```
