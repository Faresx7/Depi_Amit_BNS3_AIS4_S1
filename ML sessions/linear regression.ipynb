```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([2,2.8,3.6,4.5])
```

```python
w=0
b=0
alpha=0.01
num_iteration=20
```

```python
sse_vlaues=[]

for i in range(num_iteration):
    y_hat=w*x+b
    D_W=2/4 * np.sum((y_hat-y)*x)
    D_b=2/4 * np.sum((y_hat-y)*1)

    w-=alpha*D_W
    b-=alpha*D_b
    sse=(np.sum(y_hat-y)**2)
    sse_vlaues.append(sse)

    if(i+1)%20==0:
        print(f'iteration {i+1}, sse {sse}')
```

```python
plt.figure(figsize=(12,5))
plt.plot(range(num_iteration),sse_vlaues,label='sse')

```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y,color='red',label='Data points')
plt.plot(x,w*x+b,color='red',label='regression')
```
