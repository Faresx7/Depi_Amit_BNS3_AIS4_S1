```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([2,2.8,3.6,4.5])
```

```python
w=0
b=0
alpha=0.01
iterations=20
```

```python
sse_values=[]

for i in range(iterations):
    y_hat=w*x+b

    w_D= 2/len(x) * np.sum((y_hat-y)*x)
    b_D= 2/len(x) * np.sum(y_hat-y)

    w -= alpha*w_D
    b -= alpha*b_D

    sse=(np.sum((y_hat-y)**2))
    sse_values.append(sse)
    if (i+1)%20 ==0:
        print(f'iteration{i+1}  ,  mse{sse}\nw={w}  ,  b={b}\nD_w={w_D}  ,  D_b={b_D}')


    
```

```python
plt.plot(range(iterations),sse_values,label='sse')
```
