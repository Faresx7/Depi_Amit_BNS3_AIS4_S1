```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Linear_regression:
    #! more common default values 
    def __init__(self, alpha=0.01, num_iterations=100):
        self.alpha = alpha
        self.num_iterations = num_iterations 
        self.w= 0
        self.b= 0
        self.sse_values = []


    def fit (self , x , y):
        x =np.array(x) 
        y =np.array(y)  

        for i in range (self.num_iterations):
            y_hat = self.w * x + self.b

            D_w = (2/len(x)) * np.sum((y_hat - y) * x)
            D_b = (2/len(x) * np.sum(y_hat - y))

            self.w -=self.alpha * D_w
            self.b -=self.alpha * D_b

            sse = np.sum((y_hat - y) **2)
            self.sse_values.append(sse)

            if (i+1)%20==0:
                print(f"iteration {i+1}, SSE: {np.round(sse,4)}, w: {np.round(self.w,4)}, b: {np.round(self.b,2)}")
    
    
    def predict(self , x):
        return self.w * np.array(x) + self.b    


    def plot_regression(self, x, y):
        plt.figure(figsize=(8,2))
        plt.scatter(x,y, label="Data points")
        plt.plot(x, self.predict(x),label="Regression line")
        #to display every object name
        plt.legend()
        plt.show()


    def sse(self):
        plt.figure(figsize=(10,5))
        plt.plot(range(self.num_iterations), self.sse_values, label="SSE")
        plt.xlabel("Iteration")
        plt.ylabel("SSE")
        plt.show()         
                
```

testing code

```python
x = np.array([1,2,3,4])
y = np.array([2,2.8,3.6,4.5])
model = Linear_regression(alpha=0.01, num_iterations=20)

model.fit(x, y)
model.plot_regression(x, y)
model.sse()
```
